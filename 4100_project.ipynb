```python
# Creating clusters of the countries and grouping. 
# Before doing so find the ideal number of clusters based on the relatives death ratio over the population

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# calculate the death-to-population ratio
df['death ratio'] = df['total_death']/df['Population']

# remove nan values
df.dropna(inplace=True)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# range of clusters, cant be too many only have 19 countries.
n_clusters_range = range(4, 12)

# iterate over the range and calculate the silhouette score for each
silhouette_scores = []
for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    labels = kmeans.fit_predict(df[['death ratio']])
    silhouette_scores.append(silhouette_score(df[['death ratio']], labels))

# find the index of the maximum silhouette score
max_index = np.argmax(silhouette_scores)

# print the optimal number of clusters and the corresponding silhouette score
print('Ideal # of clusters:', n_clusters_range[max_index])
print('score of silhouette model:', silhouette_scores[max_index])
```

# **Choosing the number of clusters.**

With the previous regression model proving that that data points were all over the place we wanted to group the countries that would better fit into an individual behavior model. # Distribution of Countries 

Through the dendogram we are able to provide a visual representation of the distance between the countries in relative similarity. The clustering provides us with insight on what group of countries are most similar in their behavioral patterns with consideration of all factors. 



```python

# given 4 is the ideal number of cluster we wan to create a dendogram and then list the 
# countries associated with each cluster
import pandas as pd

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster


X = df.drop('countries', axis=1).to_numpy()
Z = linkage(X, method='average')

# create a dendrogram with 4 clusters
plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode='lastp', p=4)
plt.title('Dendrogram with 4 clusters')
plt.xlabel('Countries')
plt.ylabel('Distance')
plt.show()
k = 4
clusters = fcluster(Z, k, criterion='maxclust')

# create a bar chart that shows the countries in each cluster
cluster_counts = pd.Series(clusters).value_counts()
cluster_counts.sort_index(inplace=True)
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(cluster_counts.index, cluster_counts.values)
ax.set_xticks(cluster_counts.index)
ax.set_xticklabels([f'Cluster {i}' for i in range(1, k+1)])
ax.set_ylabel('Number of countries')
ax.set_title('Countries in each cluster')
plt.show()

for i in range(1, k+1):
    print(f'Cluster {i}: {df.loc[clusters == i, "countries"].tolist()}')
```
